### Read the CSV and Perform Basic Data Cleaning

```python
# Load the data
file_path = Path('./school_dropout.csv')
df = pd.read_csv(file_path)


df = df[df['Target'].isin(['Dropout', 'Graduate'])]

df.head()
```

```python
# Replace 'dropout' with 0 and 'graduate' with 1
import numpy as np
df['Target'] = df['Target'].replace({'Dropout': 0, 'Graduate': 1})
df['Target'] = df['Target'].astype(np.int64)
df.head()
```

```python
df.dtypes
```

```python
# Create our features
X = df.drop(columns=["Target"])

X = pd.get_dummies(X, columns=["Mother's occupation", "Father's occupation", "Gender", "Nacionality", "Marital status", "Application mode"
    ])
# Create our target
y = df["Target"]
```

```python
# Check the balance of our target values
y.value_counts()

x_cols = [i for i in df.columns if i not in ('Marital Satus', 'Target')]
X = df[x_cols]
y = df['Target']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
                                                    
```

```python
# Resample the training data with the RandomUnderSampler

from imblearn.under_sampling import RandomUnderSampler
ros = RandomUnderSampler(random_state=1)
X_resampled, y_resampled = ros.fit_resample(X_train, y_train)
Counter(y_resampled)
```

```python
# Calculated the balanced accuracy score
# YOUR CODE HERE
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver='lbfgs', random_state=1)
model.fit(X_resampled, y_resampled)
```

```python

from sklearn.metrics import confusion_matrix
y_pred = model.predict(X_test)
confusion_matrix(y_test, y_pred)

```

```python
# Display the confusion matrix
# YOUR CODE HEREconfusion_matrix(y_test, y_pred)
from sklearn.metrics import balanced_accuracy_score
balanced_accuracy_score(y_test, y_pred)
```

```python
# Print the imbalanced classification report
from imblearn.metrics import classification_report_imbalanced
print(classification_report_imbalanced(y_test, y_pred))
```

```python
from imblearn.under_sampling import ClusterCentroids
cc = ClusterCentroids(random_state=1)
X_resampled, y_resampled = cc.fit_resample(X_train, y_train)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver='lbfgs', random_state=1)
model.fit(X_resampled, y_resampled)
```

```python
from sklearn.metrics import confusion_matrix
y_pred = model.predict(X_test)
confusion_matrix(y_test, y_pred)

from sklearn.metrics import balanced_accuracy_score
balanced_accuracy_score(y_test, y_pred)

from imblearn.metrics import classification_report_imbalanced
print(classification_report_imbalanced(y_test, y_pred))
```
